```python
stocks1 = pd.DataFrame({
    'symbol':['AAPL','GOOG'],
    'open':[100,200],
    'close':[110,190],
    'high':[115,210],
    'low':[95,180],
    'parsed_time':['2020-01-01','2020-01-02']
})

stocks1.to_csv('stocks1.csv', index=False)
stocks1.to_csv('stocks2.csv', index=False)

companies = pd.DataFrame({
    'symbol':['AAPL','GOOG'],
    'name':['Apple','Google']
})

companies.to_csv('companies.csv', index=False)
```

```python
stocks1 = pd.read_csv('stocks1.csv')
stocks2 = pd.read_csv('stocks2.csv')
companies = pd.read_csv('companies.csv')
```

```python
stocks1.fillna(stocks1.max(), inplace=True)
stocks2.fillna(stocks2.min(), inplace=True)
```

```python
stocks = pd.concat([stocks1, stocks2])
```

```python
df = stocks.merge(companies, on='symbol')
```

```python
df[['open','high','low','close']].mean()
df['close'].max()
df['close'].min()
```

```python
df['parsed_time'] = pd.to_datetime(df['parsed_time'])
```

```python
df['result'] = np.where(df['close'] > df['open'], 'up', 'down')
```
