```python
class HCN:
    def __init__(self, length=0, width=0):
        self.length = length
        self.width = width

    def input(self):
        self.length = float(input("Nhập chiều dài: "))
        self.width = float(input("Nhập chiều rộng: "))

    def perimeter(self):
        return 2 * (self.length + self.width)

    def area(self):
        return self.length * self.width

    def display(self):
        print("Chiều dài:", self.length)
        print("Chiều rộng:", self.width)
        print("Chu vi:", self.perimeter())
        print("Diện tích:", self.area())


r = HCN()
r.input()
r.display()
```

```python
class ThiSinh:
    def __init__(self):
        self.hoten = ""
        self.toan = 0
        self.ly = 0
        self.hoa = 0

    def input(self):
        self.hoten = input("Họ tên: ")
        self.toan = float(input("Toán: "))
        self.ly = float(input("Lý: "))
        self.hoa = float(input("Hóa: "))

    def tong_diem(self):
        return self.toan + self.ly + self.hoa

    def display(self):
        print(self.hoten, "- Tổng điểm:", self.tong_diem())


n = int(input("Số thí sinh: "))
ds = []

for _ in range(n):
    ts = ThiSinh()
    ts.input()
    ds.append(ts)

print("\nDanh sách trúng tuyển (>=20):")
for ts in ds:
    if ts.tong_diem() >= 20:
        ts.display()
```

```python
class PhanSo:
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau

    def nhap(self):
        self.tu = int(input("Tử số: "))
        self.mau = int(input("Mẫu số: "))
        if self.mau == 0:
            raise ValueError("Mẫu số không được bằng 0")

    def hop_le(self):
        return self.mau != 0

    def hien_thi(self):
        print(f"{self.tu}/{self.mau}")


ps = PhanSo()
ps.nhap()
ps.hien_thi()
```

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.data = []

    def isEmpty(self):
        return len(self.data) == 0

    def isFull(self):
        return len(self.data) == self.size

    def push(self, x):
        if not self.isFull():
            self.data.append(x)

    def pop(self):
        if not self.isEmpty():
            return self.data.pop()

    def show(self):   
        print("Nội dung stack:", self.data)
```

```python
class Date:
    def __init__(self, d=1, m=1, y=2000):
        self.day = d
        self.month = m
        self.year = y

    def display(self):
        print(f"{self.day}/{self.month}/{self.year}")

    def next_day(self):
        self.day += 1   

d = Date()
d.display()
d.next_day()
d.display()
```

```python
class Employee:
    def __init__(self, name, birth, join):
        self.name = name
        self.birth = birth
        self.join = join

    def display(self):
        print("Tên:", self.name)
        print("Ngày sinh:", end=" ")
        self.birth.display()
        print("Ngày vào công ty:", end=" ")
        self.join.display()


b = Date(1, 1, 2000)
j = Date(1, 1, 2024)
e = Employee("Nguyen Van A", b, j)
e.display()
```

```python
class Polygon:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def perimeter(self):
        return 2 * (self.a + self.b)

    def area(self):
        return self.a * self.b


class Rectangle(Polygon):
    pass


class Square(Rectangle):
    def __init__(self, a):
        super().__init__(a, a)


s = Square(5)
print("Chu vi:", s.perimeter())
print("Diện tích:", s.area())
```

```python
import math

class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y


class Ellipse(Point):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b

    def area(self):
        return math.pi * self.a * self.b


class Circle(Ellipse):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
```

```python
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c


class IsoscelesTriangle(Triangle):
    def __init__(self, a, b):
        super().__init__(a, a, b)


class EquilateralTriangle(IsoscelesTriangle):
    def __init__(self, a):
        super().__init__(a, a)
```
