```python
import numpy as np

# 1. Tạo array từ 0–9
arr = np.arange(10)
print(arr)
print("dtype:", arr.dtype)
print("shape:", arr.shape)

# 2. Tạo arr_odd và arr_even
arr_odd = arr[arr % 2 == 1]
arr_even = arr[arr % 2 == 0]

print("Odd:", arr_odd)
print("Even:", arr_even)

# 3. Thay số lẻ bằng 100
arr_update_1 = arr.copy()
arr_update_1[arr_update_1 % 2 == 1] = 100
print(arr_update_1)
```

```python
import numpy as np

arr_a = np.array([1,2,3,2,3,4,3,4,5,6])
arr_b = np.array([7,2,10,2,7,4,9,4,9,8])

# 1. Phần tử xuất hiện ở cả arr_a và arr_b
arr_c = np.intersect1d(arr_a, arr_b)
print(arr_c)

# 2. Phần tử chỉ có ở arr_a
arr_d = np.setdiff1d(arr_a, arr_b)
print(arr_d)

# 3. Lọc giá trị từ 5 đến 10
arr_e = np.array([2,6,1,9,10,3,27,8,6,25,16])
arr_f = arr_e[(arr_e >= 5) & (arr_e <= 10)]
print(arr_f)
```

```python
import numpy as np

# build the baseball dataset from existing height and weight arrays if available
try:
	baseball = list(zip(height, weight))
except NameError:
	# fallback to empty list if height/weight are not defined
	baseball = []

np_baseball = np.array(baseball)

print("shape:", np_baseball.shape)
print("dtype:", np_baseball.dtype)

# safe access for row 49
if np_baseball.shape[0] > 49:
	print(np_baseball[49])
else:
	print(f"Row 49 not available (only {np_baseball.shape[0]} rows)")

# extract weight column if present
if np_baseball.ndim == 2 and np_baseball.shape[1] > 1:
	np_weight = np_baseball[:, 1].astype(float)
else:
	np_weight = np.array([])

# safe access for row 123, column 0
if np_baseball.shape[0] > 123:
	print(np_baseball[123, 0])
elif np_baseball.shape[0] > 0:
	print("Row 123 not available; showing last row first column:", np_baseball[-1, 0])
else:
	print("np_baseball is empty")

# compute means if possible
if np_baseball.ndim == 2 and np_baseball.shape[1] > 0:
	print("mean first column:", np_baseball[:, 0].astype(float).mean())
else:
	print("No valid first column to compute mean")

if np_weight.size > 0:
	print("mean weight column:", np_weight.mean())
else:
	print("No weight column to compute mean")
```

```python
import pandas as pd

stocks1 = pd.read_csv("stocks1.csv")
stocks2 = pd.read_csv("stocks2.csv")
companies = pd.read_csv("companies.csv")

stocks = pd.concat([stocks1, stocks2])
stocks_companies = stocks.merge(companies, on='symbol')

stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
stocks_companies['result'] = stocks_companies.apply(
    lambda x: 'up' if x.close > x.open else 'down', axis=1)
```

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
tags = pd.read_csv("tags.csv")
links = pd.read_csv("links.csv")
```

```python
movies.isnull().sum()
ratings.isnull().sum()
tags.isnull().sum()
links.isnull().sum()
```

```python
tags = tags.dropna()
```

```python
ratings['datetime'] = pd.to_datetime(ratings['timestamp'], unit='s')
tags['datetime'] = pd.to_datetime(tags['timestamp'], unit='s')
```

```python
ratings[['timestamp','datetime']].head()
```

```python
movie_ratings = pd.merge(ratings, movies, on='movieId')
movie_ratings.head()
```

```python
movie_full = pd.merge(movie_ratings, tags, on=['movieId','userId'], how='left')
movie_full.head()
```

```python
movies['movieId'].nunique()
```

```python
ratings['userId'].nunique()
```

```python
ratings['rating'].mean()
```

```python
ratings.groupby('movieId').size().sort_values(ascending=False).head()
```

```python
avg_rating = ratings.groupby('movieId')['rating'].mean()
high_rating_movies = avg_rating[avg_rating >= 4]

high_rating_movies.head()
```

```python
high_rating_movies = high_rating_movies.reset_index()
high_rating_movies = pd.merge(high_rating_movies, movies, on='movieId')
high_rating_movies[['title','rating']].head()
```

```python
movies_genres = movies.copy()
movies_genres['genres'] = movies_genres['genres'].str.split('|')
```

```python
movies_genres = movies_genres.explode('genres')
movies_genres.head()
```

```python
movies_genres['genres'].value_counts()
```

```python
genre_rating = pd.merge(movies_genres, ratings, on='movieId')

genre_rating.groupby('genres')['rating'].mean().sort_values(ascending=False)
```
